## Create Dummies

```python
df_film_month = pd.get_dummies(df_features['film_month'],prefix='film_month')
df_published_month = pd.get_dummies(df_features['published_month'],prefix='published_month')

df_film_dow = pd.get_dummies(df_features['film_dayofweek'],prefix='film_dow')
df_published_dow = pd.get_dummies(df_features['published_dayofweek'],prefix='published_dow')
```

## Drop Columns

```python
df_features.drop(columns=['film_month','published_month','film_dayofweek','published_dayofweek'], inplace=True)
```

## Join Dummies

```python
df_features = df_features.merge(df_film_month, left_index=True, right_index=True)
df_features = df_features.merge(df_published_month, left_index=True, right_index=True)
df_features = df_features.merge(df_film_dow, left_index=True, right_index=True)
df_features = df_features.merge(df_published_dow, left_index=True, right_index=True)
```

<div class="alert alert-block alert-info">

# Train, Test, Split

</div>

```python
X = df_features.drop(columns=['views'])
y = df_features['views']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Estandarizar

</div>

```python
scaler = StandardScaler()

# Scale: Fit
print(scaler.fit(X_train))

# Scale: Train Transform
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

<div class="alert alert-block alert-info">

# Red Neuronal

</div>

```python
from keras import models, layers
from keras.layers import Dropout
from keras import optimizers
from keras.layers.normalization import BatchNormalization
```

```python
def build_model(input_shape=(None,)):
    
    model = models.Sequential()
    
    # Input Layer
    model.add(layers.Dense(1024, activation='relu', input_shape=input_shape))
    
    # Hidden Layers
    model.add(layers.Dense(512, activation='relu'))
    #model.add(BatchNormalization())
    model.add(Dropout(0.1))
    model.add(layers.Dense(256, activation='relu'))
    #model.add(BatchNormalization())
    model.add(Dropout(0.1))
    model.add(layers.Dense(64, activation='relu'))
    #model.add(BatchNormalization())
    model.add(Dropout(0.1))
    model.add(layers.Dense(32, activation='relu'))
    #model.add(BatchNormalization())
    model.add(Dropout(0.1))
    
    # Output Layer
    model.add(layers.Dense(1))

    
    # Compilamos el modelo con RMSprop, pérdida MSE y métrica MAE.
    model.compile(optimizer=optimizers.RMSprop(lr=0.01), loss='mse', metrics=['mae'])
    
    return model
```

```python
num_epochs = 300
batch_size = 64

model = build_model(input_shape=(X_train.shape[1],))

history = model.fit(X_train, y_train, validation_split=0.2,  epochs=num_epochs, batch_size=batch_size, verbose=1)
```

```python
history.model.summary()
```

```python
history.history.keys()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(range(1, len(history.history['loss']) + 1), history.history['mean_absolute_error'])
#plt.plot(range(1, len(history.history['loss']) + 1), history.history['loss'])
plt.plot(range(1, len(history.history['loss']) + 1), history.history['val_mean_absolute_error'])
#plt.plot(range(1, len(history.history['loss']) + 1), history.history['val_loss'])
plt.xlabel('Epochs')
plt.ylabel('Train MAE')
plt.show()
```

```python
test_mse, test_mae = model.evaluate(X_test, y_test, verbose=1)
```
